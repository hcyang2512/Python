```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import talib
import statsmodels
%matplotlib inline
```

```python
DF=pd.read_excel(r"C:\Users\yanghch\Google 雲端硬碟\匯率.xlsx")
DF
```

```python
DF['DATE']=pd.to_datetime(DF.DATE,format='%Y%m')
DF=DF.set_index(DF['DATE'],drop=True)
del DF['DATE']
DF.head()
```

```python
region = ['USD','EUR','GBP','ZAR','JPY']
l=0
l1=1
ER ={}
for i in region:
    
    ER[i+'_B'] = DF.iloc[:,l]
    ER[i+'_S'] = DF.iloc[:,l1]
    l=l+2
    l1=l1+2
    print(ER[i+'_B'].head(),ER[i+'_S'].head())
```

```python
for i in region:   
    ER['d_'+i] = ER[i+'_S'] - ER[i+'_B']
    print(ER['d_'+i])
```

```python
close=ER['USD_S']
```

```python
#平均移动均线的指标SMA
#不同權重，MA_Type: 0=SMA, 1=EMA, 2=WMA, 3=DEMA, 4=TEMA, 5=TRIMA, 6=KAMA, 7=MAMA, 8=T3 (Default=SMA)
SMA = talib.MA(close,12,matype=0)[-1]
EMA = talib.MA(close,12,matype=1)[-1]
WMA = talib.MA(close,12,matype=2)[-1]
DEMA = talib.MA(close,12,matype=3)[-1]
TEMA = talib.MA(close,12,matype=4)[-1]
plt.scatter('SMA',SMA)
plt.scatter('EMA',EMA)
plt.scatter('WMA',WMA)
plt.scatter('DEMA',DEMA)
plt.scatter('TEMA',TEMA)
```

```python
#macd(DIF) = 12 天 EMA - 26 天 EMA， macdsignal(MACD) = 9 天 MACD的EMA ， macdhist(OSC) = MACD - MACD signal
macd, macdsignal, macdhist = talib.MACD(close, fastperiod=4, slowperiod=8, signalperiod=3)
plt.plot(macd)
plt.plot(macdsignal)
plt.plot(macdhist)
```

```python
#交易策略
record=pd.Series('NA',index=pd.date_range('1/1/1980',periods=len(macd), freq='M'))
pre=pd.Series('NA',index=pd.date_range('1/1/1980',periods=len(macd), freq='M'))
i=0
for i in range(len(macd)):
    if macd[i]>macdsignal[i] and macd[i]>0:
        record[i]=close[i]
        pre[i]='Buy'
    if macd[i]<macdsignal[i] and macd[i]<0:
        record[i]=close[i]
        pre[i]='Sell'
        i=i+1
print(record, pre)
```

```python
df2={'price':record, 'move':pre}
df2=pd.DataFrame(df2)
df2
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
from datetime import datetime

%matplotlib inline
er='https://rate.bot.com.tw/xrt'
```

```python
table = pd.read_html(er)[0]

table = table.drop(table.columns[5:12],axis=1)
table = table.drop(table.columns[5],axis=1)
```

```python
table
```

```python
#重新命名欄位名稱
table.columns = [u'幣別', u'現金_買入', u'現金_賣出', u'即期_買入', u'即期_賣出']
table
```

```python
table[u'幣別'] = table[u'幣別'].str.extract('\((\w+)\)') #幣別值有重複字 利用正規式取英文代號 ('(\w+)') 為以中文取代
```

```python
table
```

```python
#將資料存到Sqlite

#新增日期欄位&使用今天的日期
table['Date'] = datetime.now().strftime('%Y-%m-%d')
#檢查格式
table.info()
```

```python
#將date轉成pandas日期資料型式
table['Date']=pd.to_datetime(table['Date'])
table
```

```python
#存檔 以
with sqlite3.connect('currency.sqlite') as db:
    table.to_sql('currency',con=db,if_exists='append')
```

```python
with sqlite3.connect('currency.sqlite') as db:
    df=pd.read_sql_query('select * from currency', con=db)
df
```
